```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as  sns
sns.set()
```

```
dataframe = pd.read_csv(path + '/covid_19_data.csv')
covid_confirmed = pd.read_csv(path + '/time_series_covid_19_confirmed.csv')
covid_deaths = pd.read_csv(path + '/time_series_covid_19_deaths.csv')
covid_recovered = pd.read_csv(path + '/time_series_covid_19_recovered.csv')
```

```
dataThai = dataframe[dataframe['Country/Region']=='Thailand']
dataThai
```

```
dataThai['Confirmed'] = dataThai['Confirmed'].astype(float)
dataThai['Recovered'] = dataThai['Recovered'].astype(float)
```

จำนวนผู้ติดเชื้อในประเทศไทยตั้งแต่วันที่ 22/1/2563-25/2/2563

```
softfig = plt.figure(figsize=(30,5))
Graph = softfig.add_subplot(1,1,1)
x = range(0, len(dataThai['ObservationDate']))
y = dataThai['Confirmed']

Graph.plot(x,y,marker="o", color="aqua",linewidth=3)
Graph.legend(loc='best')
Graph.set_title("Confirmed Covid-19 in Thailand",color = 'indigo')
Graph.set_xlabel("Date",color='magenta')
Graph.set_ylabel("Confirmed",color='magenta')
Graph.set_xticklabels(dataThai['ObservationDate'],rotation = 45)

plt.xticks(np.arange(0, len(dataThai['ObservationDate'])))
plt.yticks(np.arange(0, 45, 5))
```

การเปรียบเทียบของผู้ติดเชื้อระหว่าง Hebei กับ Hunan ตั้งแต่วันที่ 22/1/2563-25/2/2563

```
dataHe = dataframe[dataframe['Province/State']=='Hebei']
dataHe
```

```
dataHu = dataframe[dataframe['Province/State']=='Hunan']
dataHu
```

```
softfig = plt.figure(figsize=(30,5))
Graph = softfig.add_subplot(1,1,1)
x = range(0, len(dataHe['Province/State']))
y = dataHe['Confirmed']

x1 = range(0, len(dataHu['Province/State']))
y1 = dataHu['Confirmed']

Graph.plot(x,y,marker="o", color="darkturquoise",linewidth=3,label = "Hebei")
Graph.plot(x1,y1, color="red",marker="o",linewidth=3,label="Hunan")
Graph.set_title("Compare Confirmed Covid-19 Hebei and Hunan",color = 'indigo')
Graph.set_xlabel("Date",color='magenta')
Graph.set_ylabel("Confirmed",color='magenta')
Graph.set_xticklabels(dataHe['ObservationDate'],rotation = 45)
Graph.legend(loc='best')
plt.xticks(np.arange(0, len(dataHe['ObservationDate'])))
```

```
import mpl_toolkits
from mpl_toolkits.basemap import Basemap

plt.figure(figsize=(15,10))
m = Basemap(projection='robin',lon_0=0,resolution='c')

#set a background colour
m.drawmapboundary(fill_color='#85A6D9')
m.fillcontinents(color='white',lake_color='#85A6D9')
m.drawcoastlines(color='#6D5F47', linewidth=.4)
m.drawcountries(color='#6D5F47', linewidth=.4)

Lat = covid_deaths['Lat'].values
Long = covid_deaths['Long'].values


x,y = m(Long,Lat)


m.scatter(
    x,
    y,
   
    c='lime', #color
    marker='o', #symbol
    alpha=0.5, #transparency
    zorder = 2, #plotting order

    )
```
